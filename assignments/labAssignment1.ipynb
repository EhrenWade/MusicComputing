```python
def plotAudio(file,samplingRate):
  sig , sr = librosa.load(file,mono=True,sr=None)

  time=np.arange(0,len(sig))/sr
  plt.plot(time,sig)
  plt.title('Time-Domain (Waveform)')
  plt.xlabel('Time (s)')
  plt.ylabel('Amplitude')
  plt.show()

excursion = plotAudio(file,samplingRate)
```
